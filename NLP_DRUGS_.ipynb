## Extract the data from the Drugs name

```python
# Function 1: Drug Data Extraction
def drug_data_extraction(interventions, drug_keywords_list):
    # Fill NaN values in 'name' column with an empty string
    interventions['name'] = interventions['name'].fillna('')

    # Create a regex pattern for drug keywords
    drug_pattern = '|'.join(rf'\b{re.escape(dk)}\b' for dk in drugs_keywords_list)

    # Create a boolean mask for conditions containing drug keywords
    condition_drug_mask = interventions['name'].str.contains(drug_pattern, case=False)

    # Apply the mask to filter the dataframe
    output_drug_data_ext = interventions[condition_drug_mask]

    return output_drug_data_ext



# Function to extract drugs along with dosages from text using spaCy and drug_named_entity_recognition
def extract_drugs_with_dosages(text):
    doc = Assign(text)
    drug_entities = find_drugs([t.text for t in doc], is_ignore_case=True)

    drug_info_list = []

    for entity in drug_entities:
        drug_name = entity[0]['name']
        
        # Check if dosage information is present in the entity
        if 'dosage' in entity[0]:
            dosage = entity[0]['dosage']
            drug_info_list.append(f"{drug_name} ({dosage})")
        else:
            drug_info_list.append(drug_name)

    return ', '.join(drug_info_list)

# Function# 2  for drug data extraction with desire information
def drug_data_extraction_merged(conditions, subset_for_desired_drug, studies):
    # Merging the dataframes and obtaining the disease data
    merged_result_2 = pd.merge(
        pd.merge(conditions, subset_for_desired_drug, on='nct_id', how="left", suffixes=('Disease', 'Inter')),
        studies, on='nct_id', how="left", suffixes=('', 'Studies')
    )

    # Now filtering the DRUGS column
    drug_mask = merged_result_2["intervention_type"] == "Drug"
    final_df = merged_result_2[drug_mask]
    result_df = final_df[['nct_id', 'brief_title', 'nameDisease', 'description']].tail(50)

    # Handling NaN values and converting to string
    result_df.loc[:, "description"] = result_df["description"].astype(str)
    result_df.loc[:, "brief_title"] = result_df["brief_title"].astype(str)

    # Extracting drug names with dosages from intervention description "description" column
    result_df["Drug Name"] = result_df.apply(
        lambda row: extract_drugs_with_dosages(row['description'] + ' ' + row['brief_title']), axis=1
    )

    result_df.columns = ['NCT Number', 'Study Title', 'Disease', 'Study detail (Actual)', 'Drug Name']

    return result_df

```

```python
drugs_keywords_list = []

while True:
    user_input_drugs_keywords = input("Enter the drugs and all its keywords (enter 'stop' to finish): ")
    
    if user_input_drugs_keywords.lower() == "stop":
        break
    
    drugs_keywords_list.append(user_input_drugs_keywords)
subset_for_desired_drug = drug_data_extraction(interventions, drugs_keywords_list)
drug_detail = drug_data_extraction_merged(conditions, subset_for_desired_drug, studies)
drugname=input("enter the file name")
drug_detail.to_csv(drugname+".csv") 
drug_detail
                                               
```

## Extract the data of drugs from the Disease name

```python
def disease_data_extraction(conditions, disease_keywords_list):
    # Create a case-insensitive regex pattern for disease keywords
    disease_pattern = re.compile(r'\b(?:' + '|'.join(map(re.escape, disease_keywords_list)) + r')\b', flags=re.IGNORECASE)
    
    # Use the regex pattern to filter the DataFrame
    condition_mask = conditions['name'].str.contains(disease_pattern)
    output_disease_data = conditions[condition_mask]
    
    return output_disease_data

# Function to extract drugs from text using spaCy and drug_named_entity_recognition
def extract_drugs_from_text(text):
    doc = Assign(text)
    drug_entities = find_drugs([token.text for token in doc], is_ignore_case=True)
    unique_drug_names = set(entity[0]['name'] for entity in drug_entities)
    return ', '.join(unique_drug_names)

# Function for drug data extraction
def drug_data_extraction(interventions, disease_data, studies):
    # Merge the dataframes and obtain the disease data
    merged_result = pd.merge(
        pd.merge(disease_data, interventions, on='nct_id', how="left", suffixes=('Disease', 'Inter')),
        studies, on='nct_id', how="left", suffixes=('', 'Studies')
    )

    # Filter the DRUGS column
    drug_mask = merged_result["intervention_type"] == "Drug"
    final_dataframe = merged_result[drug_mask]
    formatted_df = final_dataframe[['nct_id', 'brief_title', 'nameDisease', 'description']].head(50)

    # Handling NaN values and converting to string
    formatted_df["description"] = formatted_df["description"].astype(str)
    formatted_df["brief_title"] = formatted_df["brief_title"].astype(str)

    # Extract drug names from intervention description "description" column
    formatted_df["Extracted Drug Name"] = formatted_df.apply(lambda row: extract_drugs_from_text(row['description'] + ' ' + row['brief_title']), axis=1)

    formatted_df.columns = ['NCT Number', 'Study Title', 'Condition', 'Intervention Description', 'Extracted Drug Name']

    return formatted_df


```

```python
disease_keywords_list = []

while True:
    user_input_disease_keywords = input("Enter the disease and all its keywords (enter 'stop' to finish): ")
    
    if user_input_disease_keywords.lower() == "stop":
        break
    
    disease_keywords_list.append(user_input_disease_keywords)

disease_data=disease_data_extraction(conditions, disease_keywords_list)
df_drug_extraction = drug_data_extraction(interventions, disease_data, studies)
diseasename=input("enter the file name")
df_drug_extraction.to_csv(diseasename+".csv") 
df_drug_extraction
```
